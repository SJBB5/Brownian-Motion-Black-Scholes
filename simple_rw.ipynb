We can generally define a random walk as 

$$
S_{n+1} = S_n + X_n, \ \   n = 0,1,...
$$
where $X_i$ are iid random variables. The simplest manifestation of this process can be implemented by using a Bernoulli distribution with $p = 0.5$ for the $X_i$.\\
\\
Another slightly more complicated way is to define a sample space for movement as
$$
C = \{L_1, L_2\}
$$
Where $L_1$ and $L_2$ correspond to traveling a certain distance, and there are $p_1$ and $p_2$, which are the probabilities of traveling the corresponding distances and $p_1 + p_2 = 1$. We can derive,
$$
\begin{aligned}
E[X_i] &= L_1 p_1 + L_2 p_2 \\
\operatorname{Var}[X] &= p_1 p_2 (L_1 - L_2)^2
\end{aligned}
$$
First, we will simulate the simplest case where $L_1 = -1$ and $L_2 = 1$, with equal (p=0.5) probability.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd

n_paths = 25
n_steps = 100

rng = np.random.default_rng()

steps = rng.choice([-1.0, 1.0], size=(n_paths, n_steps)) # choice uses default of p = 0.5
paths = np.c_[np.zeros(n_paths), steps.cumsum(axis=1)] # Joins starting column and cumult increase arrays

t = np.arange(n_steps + 1)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t, paths.T, linewidth=1) 
```

Now we will define a function and simulate the case where $L_1$, $L_2$ $p_1$, and $p_2$ differ.

```python
def simulate_walks(L1, L2, p, num_paths, num_steps):
    
    rng = np.random.default_rng()
    
    steps = rng.choice([-1.0, 1.0], p =  [p, 1-p], size=(n_paths, n_steps))

    paths = np.c_[np.zeros(n_paths), steps.cumsum(axis=1)]

    t = np.arange(n_steps + 1)
    
    #plot
    fig,ax = plt.subplots(figsize = (10,6))
    ax.plot(t, paths.T, linewidth = 0.8)
    ax.set_title(f'Random Walks for L = {L1, L2} and p = {p}')
    ax.grid(True, linestyle='--', alpha=0.7) 
    
    
```
