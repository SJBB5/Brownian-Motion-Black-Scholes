Brownian motion can be thought of as a continuous version of the idea of random walk explored in simple_rw

We can define the random walk as
$$
W_n = \sum_{i=1}^n X_i,
$$
where $X_i$ are the discrete random walk steps with $\mathbb{E}[X_i]=0$ and $\mathrm{Var}(X_i)=1$.
As $n$ increases, the variance of $W_n$ grows linearly in $n$.

To prevent the variance from exploding, we introduce a scaling by $\sqrt{n}$ and define the scaled random walk:
$$
W_n(t) = \frac{1}{\sqrt{n}} \sum_{i=1}^{\lfloor nt \rfloor} X_i
$$

As $n \to \infty$, the distribution of $W_n(t)$ converges to a normal distribution with mean $0$ and variance $t$:
$$
W_n(t) \xrightarrow{d} \mathcal{N}(0, t).
$$


```python
import math
import itertools
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
M = 10 # number of simulations
t = 10 # Time variable
n = 100 # how many steps we will take in that time
dt = t/n # how large each step is
```

```python
steps = np.random.normal(0, np.sqrt(dt), size = (M,n)).T
origin = np.zeros((1, M))
bm_paths = np.concatenate([origin, steps]).cumsum(axis=0)
```

```python
time = np.linspace(0,t,n+1)
tt = np.full(shape=(M,n+1), fill_value = time).T

plt.plot(tt, bm_paths)
plt.xlabel("Years (t)")
plt.ylabel("Move")
plt.grid(True)
plt.show()
```
